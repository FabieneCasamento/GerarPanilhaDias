```python
import pandas as pd

# Crie um intervalo de datas de 01/01/1960 a 31/12/2023
data_inicial = '1960-01-01'
data_final = '2023-12-31'
dados_diarios = pd.date_range(start=data_inicial, end=data_final, freq='D')
```

```python
dados_diarios_df = pd.DataFrame({'data': dados_diarios})
```

```python
display(dados_diarios_df)
```

```python
dados_horarios = pd.date_range(start=data_inicial, end=data_final, freq='H')
```

```python
dados_horarios_df = pd.DataFrame({'data': dados_horarios})
```
